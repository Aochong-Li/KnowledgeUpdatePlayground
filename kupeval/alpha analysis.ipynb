```python
def compute_accuracy (df: pd.DataFrame, sft: bool = False):
    score = df['answer'] == df['model_answer']
    
    if not sft:
        return score.sum() / len(score)
    else:
        train_entities = pd.read_pickle('/share/goyal/lio/knowledge_delta/dataset/sft/alpha_train_test/train_sft.pickle')
        train_score = score[df['entity_id'].isin(train_entities['entity_id'])]
        test_score = score[~df['entity_id'].isin(train_entities['entity_id'])]
        
        return {'SFT Train Accuracy': train_score.sum() / len(train_score), 'SFT Test Accuracy': test_score.sum() / len(test_score)}
```

```python
def response_rate (df: pd.DataFrame):
    rate = len(df[df['model_answer'].notnull()]) / len(df)
    
    return f'Response Rate: {100 * rate} %'
```

```python
def compute_dual_accuracy (model_name: str, setting: str, pt_model: str = 'llama8b'):    
    
    # load df
    output_dir = "/share/goyal/lio/knowledge_delta/evaluation/temporal_mcq/alpha/answers/"
    df = pd.read_pickle(os.path.join(output_dir, f"{model_name}/{setting}_results.pickle"))
    is_sft = False # True if 'sft' in model_name else False
    
    # remove entities that base model fail to answer
    model_df = pd.read_pickle(f"/share/goyal/lio/knowledge_delta/evaluation/temporal_mcq/alpha/answers/{pt_model}_base/zeroshot_results.pickle")
    entity_ids = model_df[model_df['model_answer'] != model_df['fact_answer']]['entity_id'].unique()    
    df = df[~df['entity_id'].isin(entity_ids)]

    fact_score = df['fact_answer'] == df['model_answer']
    update_score = df['update_answer'] == df['model_answer']
    
    # compute accuracy
    if not is_sft:
        return {
            'update answer': update_score.sum() / len(update_score),
            'fact answer': fact_score.sum() / len(fact_score)
        }
    
    else:
        train_entities = pd.read_pickle('/share/goyal/lio/knowledge_delta/dataset/sft/alpha_train_test/train_sft.pickle')

        fact_train_score = fact_score[df['entity_id'].isin(train_entities['entity_id'])]
        fact_test_score = fact_score[~df['entity_id'].isin(train_entities['entity_id'])]
        
        update_train_score = update_score[df['entity_id'].isin(train_entities['entity_id'])]
        update_test_score = update_score[~df['entity_id'].isin(train_entities['entity_id'])]
        
        return {
            'train update answer': update_train_score.sum() / len(update_train_score),
            'train fact answer': fact_train_score.sum() / len(fact_train_score),
                
            'test update answer': update_test_score.sum() / len(update_test_score),
            'test fact answer': fact_test_score.sum() / len(fact_test_score)
        }
```

```python
def ppl_tempmcq_correlation (model_name: str, setting: str):
    article_ppl = pd.read_pickle(f'/share/goyal/lio/knowledge_delta/evaluation/perplexity/article/{model_name}.pickle')
    article_ppl = article_ppl.groupby('entity_id', group_keys = False)[['perplexity']].mean().reset_index()
    
    wiki_ppl = pd.read_pickle(f'/share/goyal/lio/knowledge_delta/evaluation/perplexity/wikipage/{model_name}.pickle')
    wiki_ppl = wiki_ppl.groupby('entity_id', group_keys = False)[['perplexity']].mean().reset_index()
    
    output_dir = "/share/goyal/lio/knowledge_delta/evaluation/temporal_mcq/alpha/answers"

    df = pd.read_pickle(os.path.join(output_dir, f"{model_name}/{setting}_confidence_level.pickle"))

    cpt_df, pt_df = df.merge(article_ppl, on = 'entity_id'), df.merge(wiki_ppl, on = 'entity_id')
    
    cpt_corr, pt_corr = cpt_df["update_confidence_level"].corr(cpt_df['perplexity']), pt_df["update_confidence_level"].corr(pt_df['perplexity'])
    
    return {
        "CPT Corr": cpt_corr,
        "PT Corr": pt_corr
    }
```

```python
model_settings = [
    ('mistral7b_cpt', 'zeroshot'),
    ('mistral7b_cpt', 'fewshot'),
    ('mistral7b_cpt', 'recall'),
    
    ('mistral7b_cpt_prior', 'zeroshot'),
    ('mistral7b_cpt_prior', 'fewshot'),
    ('mistral7b_cpt_prior', 'recall'),
    
    ('mistral7b_cpt_rephrase', 'zeroshot'),
    ('mistral7b_cpt_rephrase', 'fewshot'),
    ('mistral7b_cpt_rephrase', 'recall'),
    
    ('mistral7b_sft', 'zeroshot'),
    ('mistral7b_sft', 'fewshot'),
    ('mistral7b_sft', 'recall'),
    
]


# model_settings = [
#     ('llama8b_cpt', 'zeroshot'),
#     ('llama8b_cpt', 'fewshot'),
#     ('llama8b_cpt', 'recall'),
    
#     ('llama8b_cpt_prior', 'zeroshot'),
#     ('llama8b_cpt_prior', 'fewshot'),
#     ('llama8b_cpt_prior', 'recall'),
    
#     ('llama8b_cpt_rephrase', 'zeroshot'),
#     ('llama8b_cpt_rephrase', 'fewshot'),
#     ('llama8b_cpt_rephrase', 'recall'),
    
#     ('llama8b_sft', 'zeroshot'),
#     ('llama8b_sft', 'fewshot'),
#     ('llama8b_sft', 'recall'),
    
# ]

df_list = []

for model_setting in model_settings:
    model_name, setting = model_setting
    results = ppl_tempmcq_correlation(model_name, setting)

    df = pd.DataFrame(columns = ['Model', 'Setting', 'Correlation', 'Value'])
    for i in range(len(results)):
        perplexity = list(results.keys())[i]
        df.loc[i] = [model_name,setting, perplexity, results[perplexity]]
        
    df_list.append(df)


df = pd.concat(df_list, ignore_index = True)
```

```python
name_map = {
    "mistral7b_cpt" : "CPT",
    "mistral7b_cpt_prior": "Prior",
    "mistral7b_cpt_rephrase": "Reph",
    "mistral7b_sft": "SFT",
    
    "zeroshot": "0-shot",
    "fewshot": "4-shot",
    "recall" :"4-shot (MR)"
}

# name_map = {
#     "llama8b_cpt" : "CPT",
#     "llama8b_cpt_prior": "Prior",
#     "llama8b_cpt_rephrase": "Reph",
#     "llama8b_sft": "SFT",
    
#     "zeroshot": "0-shot",
#     "fewshot": "4-shot",
#     "recall" :"4-shot (MR)"
# }

df['Model'] = df['Model'].map(name_map)
df['Setting'] = df['Setting'].map(name_map)
```
